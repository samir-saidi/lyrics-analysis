```python
import os
from dotenv import load_dotenv
import lyricsgenius

# Load environment variables from .env file
load_dotenv()

# Get token
token = os.getenv("GENIUS_ACCESS_TOKEN")
print("Token loaded!")

# Initialize Genius API
genius = lyricsgenius.Genius(token)
```

```python
artist = genius.search_artist("Joanna Newsom", max_songs=1)
print(f"Found artist: {artist.name}")

song = genius.search_song("Sapokanikan", "Joanna Newsom")
if song:
    print(f"Title: {song.title}")
    print(f"Lyrics preview: {song.lyrics[:200]}...")  # First 200 chars
else:
    print("Song not found")
```
