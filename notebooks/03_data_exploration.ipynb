```python
import pandas as pd
import os

test_data_path = "../data/processed/all_songs_raw.csv"
if os.path.exists(test_data_path):
    df = pd.read_csv(test_data_path)
    print(f"Data shape: {df.shape}")
    print(f"\nColumns: {list(df.columns)}")
    print(f"\nFirst few rows:")
    display(df.head())
else:
    print(f"File not found: {test_data_path}")
```

```python
print("Basic Statistics:")

print(f"\nNumber of songs: {len(df)}")
print(f"Number of artists: {df['artist'].nunique()}")

print("\nSongs per artist:")
print(df['artist'].value_counts())

print("\nLyrics length (words):")
print(df['lyrics_length'].describe())
```
